```python
from PIL import Image, ImageSequence
import os
from datetime import datetime, timedelta

def extract_and_save_frames(gif_path, output_folder):
    # Open the GIF file
    with Image.open(gif_path) as img:
        # Iterate over each frame in the GIF
        for i, frame in enumerate(ImageSequence.Iterator(img)):
            # Create a file path for the output
            frame_path = f"{output_folder}/frame_{i+1:03d}.jpg"
            # Convert the frame to RGB mode before saving as JPEG
            frame = frame.convert("RGB")
            # Save the frame as a new image in JPEG format
            frame.save(frame_path, format='JPEG')

########################################################
#CHANGE THE NAME OF THE GIF IF NEEDED AND THE PATH
gif_path = 'data/images_enhanced/2024.gif'
output_folder = 'data/images_enhanced/2024'
extract_and_save_frames(gif_path, output_folder)
########################################################
```

```python
# Directory containing the files to rename
directory = output_folder

###################################################
#CHANGE THE START DATE IF NEEDED
start_date = datetime(2024, 1, 1)
###################################################

#get a list of the files in the order in which they apper in the directory
files = sorted(os.listdir(directory))


# Loop through each file and rename it
for i, file in enumerate(files):
    # Increment the day for each file
    new_date = start_date + timedelta(days=i)
    # Create the new file name
    # This example creates names like "01_03_2024", "02_03_2024", etc.
    new_name = new_date.strftime('%d_%m_%Y') + os.path.splitext(file)[1]
    # Full path for the original and new file names
    original = os.path.join(directory, file)
    new = os.path.join(directory, new_name)
    # Rename the file
    os.rename(original, new)
    print(f'Renamed "{file}" to "{new_name}"')
```
