# Preprocessing ARPAFVG data

```python
#Usefull imports
import numpy as np
import pandas as pd
```

```python
#Importing the dataset
data = pd.read_csv('data/arpafvg_fllbandiera.csv')
```

```python
data.head(-5)
```

```python
# Replace "-" with NA
data = data.replace('-', np.nan)
```

```python
# Data check
data.info()
```

```python
# Fill the missing values with the last valid observation (forward fill)
data = data.fillna(method='ffill')
data.iloc[92:95]
```

```python
data["Temp. min gradi C"].dtype
```

```python
# Convert object dtype columns to float 
data["Temp. min gradi C"] = data["Temp. min gradi C"].astype(float)
data["Temp. med gradi C"] = data["Temp. med gradi C"].astype(float)
data["Temp. max gradi C"] = data["Temp. max gradi C"].astype(float)
data["Umidita' min %"] = data["Umidita' min %"].astype(float)
data["Umidita' med %"] = data["Umidita' med %"].astype(float)
data["Umidita' max %"] = data["Umidita' max %"].astype(float)
data["Vento med km/h"] = data["Vento med km/h"].astype(float)
data["Vento max km/h"] = data["Vento max km/h"].astype(float)   
data["Dir. V. max gradi N"] = data["Dir. V. max gradi N"].astype(float)
data["Radiaz. KJ/m2"] = data["Radiaz. KJ/m2"].astype(float)
data["Press. med hPa"] = data["Press. med hPa"].astype(float)
```

```python
data.to_csv('data/arpafvg_fllbandiera_clean.csv', index=False)
```
