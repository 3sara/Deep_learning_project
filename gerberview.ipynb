```python
import pygrib

# Open the GRIB file
with pygrib.open('data/gerber.grb') as grb:
    # Select the first message
    msg = grb.message(1)
    
    # Print some information about the message
    print(msg)
    
    # Get data from the message
    data = msg.values
    
    #print ALL the data
    print(data)
```

```python
#print the image of the grb file
import matplotlib.pyplot as plt 
plt.imshow(data)
plt.show()  
```
